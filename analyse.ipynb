```python
from glob import glob
from urllib.parse import unquote
import json

f = open('xinwenlianbo_text_20110406-20151031.json')
db = json.load(f)
f.close()
```

```python
def filter_by_existence(word, db):   
    return [entry for entry in db if word in entry['content']]

def count_mention_by_year(word, db):
    mention_count_by_year = {}

    for entry in filter_by_existence(word, db):
        year = entry['publish_date'][:4]
        try:
            mention_count_by_year[year] += 1
        except KeyError:
            mention_count_by_year[year] = 1
    
    print('\n'+word)
    for year, count in sorted(mention_count_by_year.items()):
        print(year, count)

def count_mention_by_month(word, db):
    mention_count_by_month = {}

    for entry in filter_by_existence(word, db):
        year_month = entry['publish_date'][:6]
        try:
            mention_count_by_month[year_month] += 1
        except KeyError:
            mention_count_by_month[year_month] = 1
    
    print('')
    print(','.join(['word', 'year', 'month', 'count']))
    for year_month, count in sorted(mention_count_by_month.items()):
        print(','.join([word,year_month[:4], year_month[4:6], str(count)]))
        
keywords = ('计划生育', '二胎', '人口', '独生子女', '单独两孩', '计生委', '卫计委', '党')

for kw in keywords:
    count_mention_by_month(kw, db)
```
